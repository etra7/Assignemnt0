##Diploma percentage 

```
import seaborn as sns
import matplotlib.pyplot as plt

labels = ['No Diploma', 'Bachelor', 'Master', 'PhD']
no_diploma_count = len(data_df)-data_df.Bachelor.sum()-data_df.Masters.sum()-data_df.PhD.sum()
sizes = [no_diploma_count,
         data_df.Bachelor.sum(),
         data_df.Masters.sum(),
         data_df.PhD.sum()
        ]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')
plt.show()
```

##Most popular programming languages for data science 

```
soft_skills = data_df[['Python','R','Matlab','SQL']].sum().to_frame().reset_index()
soft_skills.columns = ['Language', 'count']
soft_skills['percentage']=soft_skills['count']*100/len(data_df)
ax = sns.barplot(x='Language', y='percentage',data=soft_skills)
ax.set(ylim=(0, 100))
```
